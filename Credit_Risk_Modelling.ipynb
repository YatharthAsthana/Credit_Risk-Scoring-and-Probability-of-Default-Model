```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

df = pd.read_csv("/content/accepted_2007_to_2018Q4.csv", low_memory=False)

print(f"Dataset Shape: {df.shape}")

df.head()
```

```python
print(f"Original shape: {df.shape}")

# Filtering for only closed loans
target_categories = ['Fully Paid', 'Charged Off', 'Default']
df = df[df['loan_status'].isin(target_categories)].copy()

# Mapping the target variable to 0 (Good) and 1 (Bad/Default)
df['target'] = np.where(df['loan_status'] == 'Fully Paid', 0, 1)

# Dropping the old loan_status column to avoid confusion
df = df.drop('loan_status', axis=1)

print(f"Shape after filtering active loans: {df.shape}")
print(df['target'].value_counts(normalize=True)) # See the percentage of defaults
```

```python
# List of post-origination columns that cause data leakage
leakage_cols = [
    'out_prncp', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv',
    'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee', 'recoveries',
    'collection_recovery_fee', 'last_pymnt_d', 'last_pymnt_amnt',
    'next_pymnt_d', 'last_credit_pull_d', 'debt_settlement_flag',
    'settlement_status', 'settlement_date', 'settlement_amount',
    'settlement_percentage', 'settlement_term'
]

# Dropping leakage columns if they exist in the dataframe
cols_to_drop = [col for col in leakage_cols if col in df.columns]
df = df.drop(columns=cols_to_drop)

print(f"Shape after removing data leakage: {df.shape}")
```

```python
# Core features available at the time of application
core_features = [
    'target',          # Our dependent variable
    'loan_amnt',       # Amount applied for
    'term',            # 36 or 60 months
    'int_rate',        # Interest rate assigned
    'emp_length',      # Employment length in years
    'home_ownership',  # Rent, Own, Mortgage
    'annual_inc',      # Annual income
    'purpose',         # Purpose of the loan (e.g., debt consolidation)
    'dti',             # Debt-to-Income ratio
    'fico_range_low',  # Lower boundary of FICO score
    'pub_rec',         # Number of derogatory public records
    'open_acc',        # Number of open credit lines
    'revol_util',      # Revolving line utilization rate
    'mort_acc'         # Number of mortgage accounts
]

# Filtering the dataframe to only include these core features
df_core = df[core_features].copy()

print(f"Final modeling dataset shape: {df_core.shape}")
```

```python
# Checking for missing values
print(df_core.isnull().sum())

# To fill employment length with the mode (most common value)
df_core['emp_length'] = df_core['emp_length'].fillna(df_core['emp_length'].mode()[0])

# To fill missing mortgage accounts with 0 (assuming null means no mortgage)
df_core['mort_acc'] = df_core['mort_acc'].fillna(0)

# To fill missing revolving utilization with the median
df_core['revol_util'] = df_core['revol_util'].fillna(df_core['revol_util'].median())

# To drop remaining rows with nulls (since our dataset is massive, losing a tiny fraction is fine)
df_core = df_core.dropna()

print(f"Cleaned dataset shape: {df_core.shape}")
```

```python
# Extracts the first continuous sequence of digits from the string
extracted_nums = df_core['emp_length'].str.extract(r'(\d+)')[0].astype(float)

# Created a boolean mask to flag rows that contain the '<' symbol
is_less_than = df_core['emp_length'].str.contains('<', na=False)

# Applying the dynamic logic: if '<' is present, subtract 1. Otherwise, keep the number.
df_core['emp_length'] = np.where(is_less_than, extracted_nums - 1, extracted_nums)

# Verifying the changes I made
print("\nSample of dynamically cleaned emp_length:")
print(df_core['emp_length'].value_counts(dropna=False).head())
```

```python
import scorecardpy as sc
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
print(f"   Training set shape: {train.shape}")
print(f"   Testing set shape: {test.shape}")
```

```python
# INFORMATION VALUE (IV) FILTERING
# -------------------------------------------------------------------------
print("\n3. Calculating Information Value (IV) to filter weak predictors...")
# This automatically evaluates the predictive power of each feature and drops weak ones
dt_s = sc.var_filter(train, y='target')
print(f"   Features retained after IV filtering: {dt_s.columns.tolist()}")
```

```python
# WEIGHT OF EVIDENCE (WoE) BINNING (CORRECTED)

print("\n4. Generating Weight of Evidence (WoE) bins...")
# To calculate bins based ONLY on the filtered features in dt_s
bins = sc.woebin(dt_s, y='target')

print("   Transforming raw data into WoE values...")
# Restricting the train and test sets to strictly the columns that passed the IV filter
retained_columns = dt_s.columns.tolist()

train_woe = sc.woebin_ply(train[retained_columns], bins)
test_woe = sc.woebin_ply(test[retained_columns], bins)

# -------------------------------------------------------------------------
# LOGISTIC REGRESSION MODELING
# -------------------------------------------------------------------------
print("\n5. Training the Logistic Regression model...")
# Separate features (X) and target (y)
y_train = train_woe.loc[:, 'target']
X_train = train_woe.drop('target', axis=1)

y_test = test_woe.loc[:, 'target']
X_test = test_woe.drop('target', axis=1)

# Training the model using the 'liblinear' solver
lr_model = LogisticRegression(solver='liblinear', random_state=42)
lr_model.fit(X_train, y_train)

print("Model successfully trained on strictly numeric WoE data!")
```

```python
# SCORECARD GENERATION

print("Generating the final Credit Scorecard")

# To create the scorecard using the trained model, the WoE bins, and the retained features
# Base score is 600, PDO is 50 (standard industry baselines)
card = sc.scorecard(bins, lr_model, X_train.columns, points0=600, pdo=50)

# Applying the scorecard to calculate final scores for the Test dataset
print("   Calculating credit scores for the test portfolio...")
test_score = sc.scorecard_ply(test[retained_columns], card)

# Merging the scores back with the target variable to analyze performance
final_portfolio = test[['target']].copy()
final_portfolio['credit_score'] = test_score['score']

print("\nSample of generated Credit Scores:")
print(final_portfolio.head())
```

```python
# RISK SEGMENTATION & PORTFOLIO ANALYSIS

print("\n8. Executing Risk Segmentation Strategy...")

# Defining the business cut-offs for our risk tiers
bins_strategy = [0, 500, 580, 650, 850]
labels_strategy = ['Tier D (Decline)', 'Tier C (High Risk)', 'Tier B (Medium Risk)', 'Tier A (Low Risk)']

# Segmenting the portfolio
final_portfolio['Risk_Tier'] = pd.cut(final_portfolio['credit_score'], bins=bins_strategy, labels=labels_strategy)

# Calculating the business metrics for each tier
portfolio_analysis = final_portfolio.groupby('Risk_Tier').agg(
    Total_Applicants=('target', 'count'),
    Total_Defaults=('target', 'sum')
).reset_index()

# This is to calculate the actual default rate percentage per tier
portfolio_analysis['Default_Rate (%)'] = (portfolio_analysis['Total_Defaults'] / portfolio_analysis['Total_Applicants']) * 100

print("\n--- Portfolio Risk Segmentation Matrix ---")
print(portfolio_analysis.to_string(index=False))
```

```python
# STEP 8: RISK SEGMENTATION (QUANTILE APPROACH)

print("\n8. Executing Risk Segmentation Strategy (Quantile Optimized)...")

# Used pd.qcut to divide the portfolio into 4 relatively balanced tiers based on score distribution
# labels must be ordered from lowest score (highest risk) to highest score (lowest risk)
final_portfolio['Risk_Tier'], bins_used = pd.qcut(
    final_portfolio['credit_score'],
    q=4, # Divide into 4 equal buckets
    labels=['Tier D (Decline)', 'Tier C (High Risk)', 'Tier B (Medium Risk)', 'Tier A (Low Risk)'],
    retbins=True
)

# To calculate the business metrics for each tier
portfolio_analysis = final_portfolio.groupby('Risk_Tier').agg(
    Total_Applicants=('target', 'count'),
    Total_Defaults=('target', 'sum'),
    Min_Score=('credit_score', 'min'),
    Max_Score=('credit_score', 'max')
).reset_index()

# To calculate the actual default rate percentage per tier
portfolio_analysis['Default_Rate (%)'] = (portfolio_analysis['Total_Defaults'] / portfolio_analysis['Total_Applicants']) * 100

print("\n--- Optimized Portfolio Risk Segmentation Matrix ---")
print(portfolio_analysis.to_string(index=False))
```

```python
# MODEL STABILITY TESTING (PSI) (CORRECTED)

print("\n10. Calculating Population Stability Index (PSI)...")

# Generate scores for the training set to compare against the test set
train_score = sc.scorecard_ply(train[retained_columns], card)

# Calculate PSI by passing the full DataFrames (train_score and test_score)
# and the original target labels (y_train and y_test)
psi_results = sc.perf_psi(
    score={'train': train_score, 'test': test_score},
    label={'train': y_train, 'test': y_test}
)

# To display the stability charts
import matplotlib.pyplot as plt
plt.show()
```

```python
import matplotlib.pyplot as plt

# MODEL PERFORMANCE EVALUATION (VERSION-SAFE)
# -------------------------------------------------------------------------
print("\n11. Evaluating Model Proficiency...")

# Getting the predicted probabilities (Probability of Default)
train_pred_probs = lr_model.predict_proba(X_train)[:, 1]
test_pred_probs = lr_model.predict_proba(X_test)[:, 1]

# Evaluating Testing Set (The most important metric)
print("\nGenerating Testing Set Performance Metrics...")
perf_test = sc.perf_eva(
    label=y_test,
    pred=test_pred_probs,
    title="Testing Set Performance"
)
plt.show()

# Evaluating Training Set (To check for overfitting)
print("\nGenerating Training Set Performance Metrics...")
perf_train = sc.perf_eva(
    label=y_train,
    pred=train_pred_probs,
    title="Training Set Performance"
)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# To generate hard binary predictions (0 = Paid, 1 = Default)
y_pred_classes = lr_model.predict(X_test)

# Calculating the individual metrics
accuracy = accuracy_score(y_test, y_pred_classes)
precision = precision_score(y_test, y_pred_classes)
recall = recall_score(y_test, y_pred_classes)
f1 = f1_score(y_test, y_pred_classes)

# Printing the summarized metrics
print("--- Standard Classification Metrics ---")
print(f"Accuracy:  {accuracy:.4f} ({accuracy * 100:.2f}%)")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# Printing the Confusion Matrix
print("\n--- Confusion Matrix ---")
print("Format: [[True Negatives (Paid), False Positives (False Alarms)]")
print("         [False Negatives (Missed Defaults), True Positives (Caught Defaults)]]")
print(confusion_matrix(y_test, y_pred_classes))

# Printing the detailed report
print("\n--- Detailed Classification Report ---")
print(classification_report(y_test, y_pred_classes))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import numpy as np

print("--- RE-TRAINING WITH CLASS WEIGHTS ---")
# Re-training the model, telling it that Class 1 (Defaults) is highly important
lr_model_balanced = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42)
lr_model_balanced.fit(X_train, y_train)

# Getting the raw probabilities instead of the hard 0/1 predictions
y_pred_probs_balanced = lr_model_balanced.predict_proba(X_test)[:, 1]

# Applying a custom Business Threshold (e.g., 50% probability is too high, let's trigger at 50% of the new balanced distribution)
# Because we used class_weight='balanced', the probabilities will shift.
# We will evaluate at the standard 0.5 for the balanced model first.
y_pred_custom = (y_pred_probs_balanced > 0.50).astype(int)

print("\n--- NEW Standard Classification Metrics ---")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_custom):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_custom):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_custom):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_pred_custom):.4f}")

print("\n--- NEW Confusion Matrix ---")
print(confusion_matrix(y_test, y_pred_custom))

print("\n--- NEW Detailed Classification Report ---")
print(classification_report(y_test, y_pred_custom))
```

```python
from sklearn.metrics import precision_recall_curve
import numpy as np

print("--- OPTIMIZING MODEL DECISION THRESHOLD ---")

# To Get the raw probabilities from your balanced model
y_pred_probs_balanced = lr_model_balanced.predict_proba(X_test)[:, 1]

# To Calculate precision, recall, and thresholds across the entire dataset
precisions, recalls, thresholds = precision_recall_curve(y_test, y_pred_probs_balanced)

# To Calculate the F1-Score for every single threshold
# Formula: 2 * (Precision * Recall) / (Precision + Recall)
# We use np.divide to safely handle any potential division by zero
f1_scores = np.divide(
    2 * (precisions * recalls),
    (precisions + recalls),
    out=np.zeros_like(precisions),
    where=(precisions + recalls) != 0
)

# Find the threshold that produced the absolute highest F1-Score
optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]
max_f1 = f1_scores[optimal_idx]

print(f"Optimal Probability Threshold found: {optimal_threshold:.4f}")
print(f"Maximum possible F1-Score: {max_f1:.4f}\n")

# Apply the new mathematically optimal threshold to your predictions
y_pred_optimal = (y_pred_probs_balanced >= optimal_threshold).astype(int)

# Print the final, optimized business metrics
print("--- FINAL OPTIMIZED METRICS ---")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_optimal):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_optimal):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_optimal):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_pred_optimal):.4f}")

print("\n--- FINAL CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_optimal))
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import numpy as np

print("--- TRAINING XGBOOST CHAMPION MODEL ---")

# Calculating the scale_pos_weight to handle the severe class imbalance
# Formula: (Count of Negative Class) / (Count of Positive Class)
# This tells XGBoost to pay way more attention to the minority 'Default' class
neg_class_count = (y_train == 0).sum()
pos_class_count = (y_train == 1).sum()
scale_weight = neg_class_count / pos_class_count

# Initializing the XGBoost Classifier
# We use a relatively low learning rate and max_depth to prevent overfitting
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    scale_pos_weight=scale_weight,
    learning_rate=0.1,
    max_depth=5,
    n_estimators=200,
    random_state=42,
    use_label_encoder=False
)

# To train the model on your WoE transformed data
xgb_model.fit(X_train, y_train)

# To predict probabilities on the Test set
y_pred_probs_xgb = xgb_model.predict_proba(X_test)[:, 1]

# To apply the standard 0.5 threshold first (XGBoost's scale_pos_weight optimizes for this)
y_pred_xgb = (y_pred_probs_xgb >= 0.50).astype(int)

# To evaluate the Champion Model
print("\n--- XGBOOST CLASSIFICATION METRICS ---")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_xgb):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_xgb):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_xgb):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_pred_xgb):.4f}")

print("\n--- XGBOOST CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_xgb))
```

## Feature Engineering

```python
import numpy as np

print("--- INITIATING ADVANCED FEATURE ENGINEERING ---")

# To prevent division by zero errors, we replace 0 income with a tiny placeholder
df_core['annual_inc'] = df_core['annual_inc'].replace(0, 0.01)

# Feature 1: Loan-to-Income Ratio (LTI)
# Why it matters: Asking for a $30k loan is very different for someone making $40k vs $150k.
df_core['loan_to_income'] = df_core['loan_amnt'] / df_core['annual_inc']

# Feature 2: Credit-worthiness Scaled by Debt (FICO-to-DTI)
# Why it matters: A high FICO score might not save you if you are drowning in a high Debt-to-Income ratio.
# We add +1 to DTI to avoid dividing by zero.
df_core['fico_to_dti'] = df_core['fico_range_low'] / (df_core['dti'] + 1)

# Feature 3: Non-Mortgage Account Concentration
# Why it matters: Having 15 credit cards (open_acc) but no mortgage implies different risk than 15 accounts including a mortgage.
df_core['credit_to_mortgage_ratio'] = df_core['open_acc'] / (df_core['mort_acc'] + 1)

print(f"Engineered 3 new financial ratios. New dataset shape: {df_core.shape}")
print(df_core[['loan_to_income', 'fico_to_dti', 'credit_to_mortgage_ratio']].head())
```

## Model Stacking Technique
## Stacking (Stacked Generalization).

The Execution: We train our WoE Logistic Regression model and Random Forest model. Instead of looking at their final outputs, we take their predicted probabilities and feed them as input features into a brand new, third model (a "Meta-Learner", usually a simple Logistic Regression). The Meta-Learner learns when to trust the linear math and when to trust the non-linear trees, generally outperforming both.

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np

print("--- 1. INITIALIZING BASE MODELS ---")
# Base Model 1: The linear, highly interpretable Logistic Regression
base_lr = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42)

# Base Model 2: The complex, non-linear Random Forest
# Kept slightly shallow (max_depth=10) to ensure it doesn't overfit before reaching the Meta-Learner
base_rf = RandomForestClassifier(n_estimators=150, max_depth=10, class_weight='balanced', random_state=42, n_jobs=-1)

print("\n--- 2. BUILDING THE STACKING ARCHITECTURE ---")
# The Meta-Learner: A simple Logistic Regression to weigh the outputs of the two base models
meta_learner = LogisticRegression(solver='liblinear', random_state=42)

# Combine them using scikit-learn's StackingClassifier
# cv=5 ensures the meta-learner trains on out-of-fold predictions to prevent massive data leakage
stacked_model = StackingClassifier(
    estimators=[
        ('Logistic_Regression', base_lr),
        ('Random_Forest', base_rf)
    ],
    final_estimator=meta_learner,
    cv=5,
    n_jobs=-1
)

print("Training the Stacking Ensemble... (Please wait, this will take a few minutes)")
stacked_model.fit(X_train, y_train)

print("\n--- 3. OPTIMIZING META-LEARNER DECISION THRESHOLD ---")
# Get raw probabilities from the Meta-Learner on the unseen Test set
y_pred_probs_stacked = stacked_model.predict_proba(X_test)[:, 1]

# Dynamically find the mathematically optimal threshold for the F1-Score
precisions, recalls, thresholds = precision_recall_curve(y_test, y_pred_probs_stacked)

f1_scores = np.divide(
    2 * (precisions * recalls),
    (precisions + recalls),
    out=np.zeros_like(precisions),
    where=(precisions + recalls) != 0
)

optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]
print(f"Optimal Probability Threshold found: {optimal_threshold:.4f}")

# Apply the optimal threshold to our predictions
y_pred_stacked = (y_pred_probs_stacked >= optimal_threshold).astype(int)

print("\n--- 4. FINAL STACKED ENSEMBLE METRICS ---")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_stacked):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_stacked):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_stacked):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_pred_stacked):.4f}")

print("\n--- FINAL CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_stacked))
```

## Portfolio Segmentation (Sub-Modeling)

```python
import scorecardpy as sc
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd

print("--- INITIATING PORTFOLIO SEGMENTATION (SUB-MODELING) ---")

# Define our segmentation strategy
segment_col = 'term'
unique_segments = df_core[segment_col].dropna().unique()

# Dictionary to elegantly store our separate models and metrics
portfolio_models = {}

for segment in unique_segments:
    print(f"\n======================================================")
    print(f"BUILDING CUSTOM SCORECARD FOR: {segment}")
    print(f"======================================================")

    # Isolate the specific business segment
    df_segment = df_core[df_core[segment_col] == segment].copy()

    # Drop the segment column itself (zero variance for this sub-population)
    df_segment = df_segment.drop(columns=[segment_col])

    print(f"Population Size: {df_segment.shape[0]} borrowers")

    try:
        # Split the segmented data (70% Train / 30% Test)
        split_data = sc.split_df(df_segment, 'target', ratio=0.7, seed=42)
        train_seg = split_data['train']
        test_seg = split_data['test']

        # Recalculate Information Value strictly for this segment
        print("Calculating segment-specific Information Value (IV)...")
        # FIX: Removed the unsupported 'return_drop' parameter
        dt_s_seg = sc.var_filter(train_seg, y='target', iv_limit=0.02)
        retained_cols_seg = dt_s_seg.columns.tolist()

        # To generate bespoke WoE bins for this specific population
        bins_seg = sc.woebin(dt_s_seg, y='target', print_info=False)

        # To apply the new WoE transformations
        train_woe_seg = sc.woebin_ply(train_seg[retained_cols_seg], bins_seg, print_info=False)
        test_woe_seg = sc.woebin_ply(test_seg[retained_cols_seg], bins_seg, print_info=False)

        y_train_seg = train_woe_seg['target']
        X_train_seg = train_woe_seg.drop('target', axis=1)

        y_test_seg = test_woe_seg['target']
        X_test_seg = test_woe_seg.drop('target', axis=1)

        # To Train the customized Champion model for this segment
        lr_model_seg = LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42)
        lr_model_seg.fit(X_train_seg, y_train_seg)

        # To evaluate the segment's performance
        y_pred_seg = lr_model_seg.predict(X_test_seg)

        print(f"\n--- METRICS FOR {segment} PORTFOLIO ---")
        print(f"Accuracy:  {accuracy_score(y_test_seg, y_pred_seg):.4f}")
        print(f"Precision: {precision_score(y_test_seg, y_pred_seg):.4f}")
        print(f"Recall:    {recall_score(y_test_seg, y_pred_seg):.4f}")
        print(f"F1 Score:  {f1_score(y_test_seg, y_pred_seg):.4f}")

        print("\n--- CONFUSION MATRIX ---")
        print(confusion_matrix(y_test_seg, y_pred_seg))

        # Save the finalized model and bins
        portfolio_models[segment] = {
            'model': lr_model_seg,
            'bins': bins_seg
        }

    except Exception as e:
        print(f"Could not build model for segment {segment}. Error: {e}")

print("\n======================================================")
print("PORTFOLIO SEGMENTATION COMPLETE")
print("======================================================")
```

```python
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("--- 1. PREPARING RAW DATA FOR CATBOOST ---")
# We MUST use the raw df_core, NOT the WoE-transformed data.
y_raw = df_core['target']
X_raw = df_core.drop('target', axis=1)

# CatBoost needs to know exactly which columns are categorical text
cat_features = X_raw.select_dtypes(include=['object', 'category']).columns.tolist()

# Filling any stray NaNs in categorical columns with a string placeholder
X_raw[cat_features] = X_raw[cat_features].fillna('MISSING')

# Splitting the raw data (70% Train, 30% Test)
X_train_cb, X_test_cb, y_train_cb, y_test_cb = train_test_split(
    X_raw, y_raw, test_size=0.3, random_state=42, stratify=y_raw
)

print(f"Categorical features natively handled by CatBoost: \n{cat_features}")

print("\n--- 2. TRAINING THE CATBOOST CHAMPION MODEL ---")
# Initializing the CatBoost model
# 'auto_class_weights=Balanced' perfectly handles our massive defaulter imbalance automatically
cb_model = CatBoostClassifier(
    iterations=500,               # Number of sequential trees to build
    learning_rate=0.05,           # Slower learning rate prevents overfitting
    depth=6,                      # Depth of trees
    cat_features=cat_features,    # Passing the raw text columns directly
    auto_class_weights='Balanced',
    eval_metric='F1',
    random_seed=42,
    verbose=50
)

# Training the model while evaluating on the test set simultaneously to prevent overfitting
cb_model.fit(
    X_train_cb, y_train_cb,
    eval_set=(X_test_cb, y_test_cb),
    early_stopping_rounds=50
)

print("\n--- 3. FINAL CATBOOST EVALUATION METRICS ---")
# Generate hard predictions on the unseen test set
y_pred_cb = cb_model.predict(X_test_cb)

# Ensure predictions are integers for scikit-learn metrics
y_pred_cb = y_pred_cb.astype(int)

print(f"Accuracy:  {accuracy_score(y_test_cb, y_pred_cb):.4f}")
print(f"Precision: {precision_score(y_test_cb, y_pred_cb):.4f}")
print(f"Recall:    {recall_score(y_test_cb, y_pred_cb):.4f}")
print(f"F1 Score:  {f1_score(y_test_cb, y_pred_cb):.4f}")

print("\n--- CATBOOST CONFUSION MATRIX ---")
print(confusion_matrix(y_test_cb, y_pred_cb))
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import precision_recall_curve, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np
import scipy.stats as stats

print("--- 1. INITIALIZING EXHAUSTIVE XGBOOST OPTIMIZATION ---")

# Calculate base weight for imbalanced classes
neg_class = (y_train == 0).sum()
pos_class = (y_train == 1).sum()
scale_weight = neg_class / pos_class

# Initialize the base XGBoost model
xgb_base = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='aucpr', # Optimizing for Precision-Recall Area Under Curve
    random_state=42,
    use_label_encoder=False
)

# Define the hyperparameter grid
# We are testing everything from shallow/fast trees to deep/complex trees
param_dist = {
    'max_depth': stats.randint(3, 10),
    'learning_rate': stats.uniform(0.01, 0.2),
    'subsample': stats.uniform(0.6, 0.4),
    'colsample_bytree': stats.uniform(0.6, 0.4),
    'n_estimators': stats.randint(100, 400),
    'scale_pos_weight': [1, scale_weight, scale_weight * 0.5] # Testing different aggression levels
}

print("Running Randomized Search... (This will test 20 different advanced model architectures)")
# Setting up Randomized Search
random_search = RandomizedSearchCV(
    xgb_base,
    param_distributions=param_dist,
    n_iter=20,
    scoring='f1',
    cv=3,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

# Training the search on the data
random_search.fit(X_train, y_train)

print(f"\nBest Model Parameters Found: {random_search.best_params_}")

# Extracting the absolute best champion model
xgb_champion = random_search.best_estimator_

print("\n--- 2. CUSTOM THRESHOLD OPTIMIZATION FOR F1 & ACCURACY ---")
# Predicting raw probabilities on the unseen Test set
y_pred_probs_champ = xgb_champion.predict_proba(X_test)[:, 1]

# Calculating metrics across all possible thresholds
precisions, recalls, thresholds = precision_recall_curve(y_test, y_pred_probs_champ)

# Calculating F1 for all thresholds safely
f1_scores = np.divide(
    2 * (precisions[:-1] * recalls[:-1]),
    (precisions[:-1] + recalls[:-1]),
    out=np.zeros_like(precisions[:-1]),
    where=(precisions[:-1] + recalls[:-1]) != 0
)

# We also want to calculate Accuracy at each threshold to find the "Sweet Spot"
# calculating accuracy for thousands of thresholds is heavy, so we sample 100 points
best_combined_score = 0
optimal_thresh = 0.5

print("Scanning for the perfect balance between Accuracy and F1...")
for thresh in np.linspace(0.1, 0.9, 100):
    temp_preds = (y_pred_probs_champ >= thresh).astype(int)
    temp_acc = accuracy_score(y_test, temp_preds)
    temp_f1 = f1_score(y_test, temp_preds)

    # Custom formula: We want to maximize BOTH.
    # If F1 gets close to 0.55+ and Accuracy stays above 70%, we win.
    combined_score = temp_acc + temp_f1

    if combined_score > best_combined_score:
        best_combined_score = combined_score
        optimal_thresh = thresh

print(f"\nOptimal Sweet-Spot Threshold Found: {optimal_thresh:.4f}")

# Apply the sweet-spot threshold
y_pred_final = (y_pred_probs_champ >= optimal_thresh).astype(int)

print("\n--- 3. FINAL MAXIMUM PERFORMANCE METRICS ---")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_final):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_final):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_final):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_pred_final):.4f}")

print("\n--- FINAL CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_final))
```
